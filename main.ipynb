## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
#your code here
    x = 1 #assign 1 because the multiplication
    for n in range(1, n + 1): # To have the valid number we must introduce the number +1 for stop in the introduced number"
        x *= n  # same as (x = x * n)
    return x
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

```python
# This will test your function 
test_factorial(factorial)
```

## 8. Write a function that returns the mode of a list, i.e.: the element that appears the most times.
`NOTE: You should not use count... üßê`

```python
def mode_counter(arr):
#your code here
    max_count = 0
    mode = None

    for name in arr:
        current_count = 0 
        
        for item in arr:
            if name == item:
                current_count += 1 
        
        if current_count > max_count:
            max_count = current_count
            mode = name 
    return mode
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. üòâ`

```python
def st_dev(list_sd):
#your code here
# 1. Initialize variables
    count = 0
    total_sum = 0
    
    # 2. First Loop: Calculate N and Sum
    for number in list_sd:
        count += 1
        total_sum += number
        
    # Safety Check: If list is empty or has 1 item, return 0
    if count <= 1:
        return 0.0
        
    # 3. Calculate Mean
    mean = total_sum / count
    
    # 4. Second Loop: Calculate Variance Sum
    variance_sum = 0
    for number in list_sd:
        diff = number - mean
        variance_sum += diff ** 2
        
    # 5. Calculate Final SD (Sample SD uses count - 1)
    variance = variance_sum / (count - 1)
    
    return variance ** 0.5
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
#your code here
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    
    # 1. Pick a letter from the alphabet that we MUST find
    for letter in alphabet:
        found = False
        
        # 2. Search for this letter inside the text
        for char in string:
            # Check both cases (lower and upper) manually
            # This logic replaces ".lower()" and "in" operator
            if char == letter or char == letter.upper():
                found = True
        
        # 3. If we finished looking through the text and didn't find the letter:
        if found == False:
            return False
            
    # 4. If we survive the whole alphabet loop, it is a pangram
    return True
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python
def sort_alpha(string):
#your code here
    # 1. Manual Split
    words = []
    current_word = ""
    
    for char in string:
        if char == ',':
            words.append(current_word)
            current_word = "" # Reset for next word
        else:
            current_word += char
            
    words.append(current_word) # Catch the last word
    
    # 2. Sort
    sorted_words = sorted(words)
    
    # 3. Manual Join
    result = ""
    is_first = True
    
    for word in sorted_words:
        if is_first == False:
            result += "," # Add comma before the word (if not first)
        
        result += word
        is_first = False # Flag that we passed the first word
        
    return result
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
#your code here
    # 1. Initialize Flags
    length = 0
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False
    
    # 2. Iterate
    for char in password:
        length += 1
        
        if 'a' <= char <= 'z':
            has_lower = True
            
        elif 'A' <= char <= 'Z':
            has_upper = True
            
        elif '0' <= char <= '9':
            has_digit = True
            
        else:
            # 3. The logic fix:
            # If it is NOT a letter and NOT a number, treat it as special.
            # This covers }, {, ], ), ^ and anything else.
            has_special = True
            
    # 4. Final Check
    if (length >= 8 and 
        has_lower and 
        has_upper and 
        has_digit and 
        has_special):
        return True
    else:
        return False
```

```python
# This will test your function 
test_pass(check_pass)
```
